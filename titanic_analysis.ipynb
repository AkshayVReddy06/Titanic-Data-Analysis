 DATA CLEANING

 In this step, we handle missing values in the dataset.
 Age column has missing values, so we fill it with median.
 Embarked column has missing values, so we fill it with mode.
 Cabin column has too many missing values, so we drop it.

```python
print("Missing Values:\n", df.isnull().sum())

df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

```python
# This analysis checks which gender had a higher survival rate.

gender_survival = df.groupby('Sex')['Survived'].mean()
print("\nSurvival Rate by Gender:\n", gender_survival)

# Insight:
# Females had a higher survival rate compared to males.
```

```python
# This analysis checks whether passenger class affected survival.

class_survival = df.groupby('Pclass')['Survived'].mean()
print("\nSurvival Rate by Class:\n", class_survival)

# Insight:
# 1st class passengers had the highest survival rate,
# while 3rd class passengers had the lowest.
```

```python
# This analysis groups passengers by age and checks survival rate.

bins = [0, 12, 20, 40, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Mid-age', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_survival = df.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival Rate by Age Group:\n", age_survival)

# Insight:
# Children had better survival chances compared to older passengers.
```

 DATA VISUALIZATION (ANIMATED GRAPHS)

 In this section, we visualize the survival patterns using graphs.
 Animation is added to make graphs more interactive and attractive

```python
# This graph shows survival rate by gender.

fig, ax = plt.subplots()

def animate_gender(i):
    ax.clear()
    values = gender_survival.values * (i / 20)
    ax.bar(gender_survival.index, values)
    ax.set_title("Survival Rate by Gender (Animated)")

ani = FuncAnimation(fig, animate_gender, frames=20, interval=100)
plt.show()

# Insight:
# The graph clearly shows that females survived more than males.
```

```python
# This graph shows survival rate by passenger class.

fig, ax = plt.subplots()

def animate_class(i):
    ax.clear()
    values = class_survival.values * (i / 20)
    ax.bar(class_survival.index.astype(str), values)
    ax.set_title("Survival Rate by Passenger Class (Animated)")

ani = FuncAnimation(fig, animate_class, frames=20, interval=100)
plt.show()

# Insight:
# 1st class passengers had higher survival compared to others.
```

```python
# This graph shows distribution of passenger ages.

fig, ax = plt.subplots()

def animate_age(i):
    ax.clear()
    ax.hist(df['Age'], bins=30)
    ax.set_title(f"Age Distribution (Frame {i})")

ani = FuncAnimation(fig, animate_age, frames=10, interval=200)
plt.show()

# Insight:
# Most passengers were between 20 and 40 years old. 
```
